#### Base

```python
import pandas as pd
import numpy as np
import os

directory = os.getcwd()

os.chdir(directory)
```

```python
path_file = 'MegaSena.xlsx'

df = pd.read_excel(path_file, sheet_name='megasena', header=0)
```

```python
df.columns
```

```python
df = df[['Concurso', 'Data', 'bola 1', 'bola 2', 'bola 3', 'bola 4', 'bola 5', 'bola 6']]
```

#### Analise Números Sorteados Mega Sena

```python
array_jogos = np.array(df[['bola 1', 'bola 2', 'bola 3', 'bola 4', 'bola 5', 'bola 6']])
```

```python
# Values counts de cada número sorteado.
valores, contagem = np.unique(array_jogos, return_counts=True)
frequencia_numeros = dict(zip(valores, contagem))
df_frequencia = pd.DataFrame(list(frequencia_numeros.items()), columns=['Número', 'Frequência']).sort_values(by='Frequência', ascending=False).reset_index(drop=True)
# Salvar em CSV
df_frequencia.to_csv('frequencia_numeros_sorteados.csv', index=False)
```

```python
display(df_frequencia)
```

#### Analise Números Escolhidos

```python
df_num_escolhidos = pd.read_csv('numeros_escolhidos.csv', sep=';', header=None)
array_jogos_esc = np.array(df_num_escolhidos)

array_jogos_esc
```

```python
# Values counts de cada número sorteado.
valores_esc, contagem_esc = np.unique(array_jogos_esc, return_counts=True)
frequencia_numeros_esc = dict(zip(valores_esc, contagem_esc))
df_frequencia_esc = pd.DataFrame(list(frequencia_numeros_esc.items()), columns=['Número', 'Frequência'])
df_frequencia_esc = df_frequencia_esc[1:].sort_values(by='Frequência', ascending=False).reset_index(drop=True)
# Salvar em CSV
df_frequencia_esc.to_csv('frequencia_numeros_escolhidos.csv', index=False)
```
