```python
#### Teste de jogos.
from itertools import combinations

numeros_fixos = [1, 3, 4, 10, 11, 13, 14, 20, 24, 25] # 5
numeros_fixos = [1, 4, 5, 7, 8, 10, 12, 20, 22, 25] 
numeros_fixos = [1, 4, 10, 11, 13, 14, 20, 24, 25]  # 811 jogos.
numeros_fixos = [1, 9, 10, 11, 12, 13, 14, 20, 24, 25]
numeros_fixos = [1, 2, 9, 10, 12, 13, 20, 24, 25]  #! MAIS TOP.
numeros_fixos = [1, 10, 20, 25]

c = 0
for jogo in array_jogos:
    if all(num in jogo for num in numeros_fixos):
        #print(jogo)
        c += 1
print(f'Total de jogos encontrados: {c}')
```

```python
##### NÚMEROS NÃO JOGAR #####
numeros_nao_jogar = [3, 5]
```

```python
#### Mostrar os jogos que contém os números fixos.
for jogo in array_jogos:
    if all(num in jogo for num in numeros_fixos):
        print(jogo)
```

```python
#### Quais são os números que mais aparecem junto com os números fixos nos jogos encontrados.
from collections import Counter

numeros_complementares = []
for jogo in array_jogos:
    if all(num in jogo for num in numeros_fixos):
        for num in jogo:
            if num not in numeros_fixos:
                numeros_complementares.append(num)
contador = Counter(numeros_complementares)
for numero, frequencia in contador.most_common():  
    print(f'Número: {numero}, Frequência: {frequencia}')
```

```python
#### Jogos que não tenham os números fixos.
total_jogos = 0
for jogo in array_jogos:
    if all(num not in jogo for num in numeros_fixos):
        print(jogo)
        total_jogos += 1
print(f'Total de jogos sem os números fixos: {total_jogos}')
```

#### Analizes de jogos feitos

```python
#### JOGOS ####
jogos = {
    'Lore': [1, 2, 4, 6, 7, 10, 11, 12, 14, 15, 18, 20, 22, 24, 25],
    'Cida': [1, 2, 4, 6, 8, 9, 10, 12, 13, 17, 19, 20, 21, 24, 25],
    'Joao': [1, 2, 4, 6, 7, 9, 10, 12, 13, 16, 18, 20, 23, 24, 25]
}

# Qtde de acertos por jogo nos últimos 100 concursos.
acertos_por_jogo = {
    'Lore': [],
    'Cida': [],
    'Joao': []
}


for jogo in array_jogos[0:50]:
    for nome, numeros in jogos.items():
        acertos = len(set(jogo) & set(numeros))
        if acertos >= 11:
            acertos_por_jogo[nome].append(acertos)

print('Qtde de jogos:', len(array_jogos[0:50]))
for nome, acertos in acertos_por_jogo.items():
    print(f'{nome}: {len(acertos)} acertos | Detalhes: {acertos}')

# Total de acertos somando todos os nomes.
total_acertos = sum([len(acertos) for nome, acertos in acertos_por_jogo.items()])
print(f'\nTotal de acertos somando todos os nomes: {total_acertos}')
```
